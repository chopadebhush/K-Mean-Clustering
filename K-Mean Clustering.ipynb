```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_blobs
X ,y =make_blobs(n_samples=1000,n_features=2,centers=3,random_state=1)
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train,y_test =train_test_split(X ,y,test_size=0.33,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans
```

```python
kmeans.fit(X_train)
```

```python
y_label = kmeans.predict(X_train)
y_label
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_label)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_train ,y_label)
```

```python
kmeans.labels_
```

```python
kmeans.inertia_
```

```python
## Select K centroid by using elbow method
wcss =[]
for i in range(1,10):
    kmeans =KMeans(n_clusters=i)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 10), wcss)
plt.xticks(range(1, 10))
plt.xlabel("No of clusters")
plt.ylabel("wcss")
plt.show()
```
